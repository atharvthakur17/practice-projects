# DATA CLEANING

```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
import string
stop=stopwords.words("english")
stop+=list(string.punctuation)
stop
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
     
```

```python
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
def clean_review(words):
    output=[]
    for i in words:
        if i.lower() not in stop:
            tag_pos=pos_tag([i])
            clean=lem.lemmatize(i,pos=get_simple_pos(tag_pos[0][1]))
            output.append(clean.lower())
    return output
```

```python
doc[:5]
```

```python
document=[(clean_review(word),category) for word,category in doc ]
```

```python
document[0]
```

we will be using top words in as features for classifier 

# split data 

```python
train_doc=document[0:1500]
test_doc=document[1500:]

train_doc[0]
```

ALL words in given data to find top words

```python
all_words=[]
for i in train_doc:
    all_words+=i[0]
all_words    
```

```python
import nltk
freq=nltk.FreqDist(all_words)
freq
```

```python
common=freq.most_common(4000)
common
```

```python
#now we have feature ---- the words which are most commonly used
features=[i[0] for i in common]
features
```

```python
#we need to convert data into a dictionry of words with its value alon with category

def get_feature_dict(words):  
    dict={}
    word_set=set(words)
    for i in features:
        dict[i]=i in word_set
    return dict
```

```python
train_data=[((get_feature_dict(doc)),category) for doc, category in train_doc]
test_data=[((get_feature_dict(doc)),category) for doc, category in test_doc]
```

# train

```python
from nltk import NaiveBayesClassifier
classifier=NaiveBayesClassifier.train(train_data)
```

```python
nltk.classify.accuracy(classifier,test_data)
```

```python
classifier.show_most_informative_features(15)
```
